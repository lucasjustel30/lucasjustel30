Terceira etapa

Ler o txt 'Base.csv'' e imprimi-la toda no xlsx ''Sumario.xlsm''



```python
import pandas as pd
from openpyxl import Workbook

# Ler o arquivo Base.csv
dados_base = pd.read_csv('Base.csv')

# Criar um novo arquivo Excel
arquivo_excel = 'Sumario.xlsm'
writer = pd.ExcelWriter(arquivo_excel, engine='openpyxl')

# Converter os dados em um DataFrame do pandas
dados_dataframe = pd.DataFrame(dados_base)

# Escrever o DataFrame no arquivo Excel
dados_dataframe.to_excel(writer, sheet_name='Nome da Planilha', index=False)

# Salvar as alterações no arquivo Excel
writer.save()



```
